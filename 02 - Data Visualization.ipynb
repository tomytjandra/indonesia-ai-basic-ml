## Types of Visualization

<img src='assets/chart-suggestions.jpg'>

Four categories of chart:

1. Distribution
2. Comparison
3. Relationship
4. Composition

Based on number of variables, visualization can be breakdown into three categories:

1. Single Variable
    - Displot (distribution): center, spread, skewness, outliers, multiple modes
    - Boxplot (quartiles): median, IQR, outliers, range
    - Stacked bar: comparison between categories


2. Two Variables
    - Scatter plot: numeric vs numeric, correlation (linear relationship)
    - Bar plot: category vs numeric


3. Multi Variables
    - Heatmap: matrix represented as colors, helpful to represent correlation

# Load Library

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt # plotting
import seaborn as sns

plt.style.use('seaborn')
```

# Iris Datasets

Iris data reference: https://archive.ics.uci.edu/ml/datasets/iris

## Load Data

```python
iris = pd.read_csv("datasets/iris.data",
                   names=['sepal_length',
                          'sepal_width',
                          'petal_length',
                          'petal_width',
                          'species'])
iris.head()
```

## Inspect Data

```python
print(f"Rows: {iris.shape[0]}")
print(f"Columns: {iris.shape[1]}")
```

```python
iris['species'].unique()
```

```python
iris.isnull().any()
```

```python
iris.describe()
```

## Bar Chart (Comparison)

```python
iris['species'].hist()
plt.show()
```

Insight: Ketiga species memiliki banyak row yang sama, yaitu 50.

## Scatter Plot (Relationship)

```python
iris.plot(kind='scatter',
          x='sepal_length',
          y='sepal_width')
plt.show()
```

```python
iris.plot(kind='scatter',
          x='petal_length',
          y='petal_width')
plt.show()
```

Insight:
- Sepal length dan width hampir tidak memiliki korelasi.
- Petal length dan width memilki korelasi yang positif.

## Distplot (Distribution)

`bins` semakin besar semakin mendetail

```python
sns.distplot(iris['sepal_length'], bins=20)
plt.show()
```

## Box plot (Distribution)

```python
sns.boxplot(data=iris)
plt.xlabel('Features')
plt.ylabel('Length in cm')
plt.show()
```

Insight:
- Dapat melihat nilai min, max, dan mediannya
- Pada `sepal_width` terdapat outlier

```python
sns.boxplot(x='species',
            y='petal_length',
            data=iris)
plt.xlabel('Species')
plt.ylabel('Petal Length in cm')
plt.show()
```

Insight: range `petal_length` untuk setiap `species` sangatlah berbeda.

## Pair plot

Semua kolom pada dataset `iris` langsung disajikan dalam 1 visualisasi.

```python
sns.pairplot(data=iris, hue='species')
plt.show()
```

## Heatmap Correlation

```python
sns.heatmap(data=iris.corr(),
            annot=True,
            cmap='RdBu')
plt.show()
```
