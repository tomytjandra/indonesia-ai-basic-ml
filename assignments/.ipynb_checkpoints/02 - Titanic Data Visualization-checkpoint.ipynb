# Import Libraries

Mari kita memanggil library yang akan digunakan pada notebook ini.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
```

# Data Loading

Load data berupa `csv` file dari folder `datasets`.

```python
titanic = pd.read_csv("../datasets/titanic.csv")
titanic.head()
```

```python
print(f"Rows: {titanic.shape[0]}")
print(f"Columns: {titanic.shape[1]}")
```

Terdapat 887 baris dan 8 kolom pada data `titanic`, dengan data deskripsi sebagai berikut:

1. `Survived`: Status survival penumpang (0 = Not Survived, 1 = Survived)
2. `Pclass`: Kelas tiket (1, 2, 3)
3. `Name`: Nama penumpang
4. `Sex`: Jenis kelamin penumpang
5. `Age`: Usia penumpang
6. `Siblings/Spouses Aboard`: Banyaknya saudara atau pasangan yang ikut berlabuh
7. `Parents/Children Aboard`: Banyaknya orang tua atau anak yang ikut berlabuh
8. `Fare`: Harga tiket

# Data Preprocessing

Sebelum melangkah ke tahap visualisasi, ada baiknya kita melakukan pembersihan data untuk memastikan bahwa data sudah siap.

## Check Missing Values

Cek kelengkapan nilai pada data. Ternyata data yang kita miliki sudah lengkap.

```python
titanic.isnull().any()
```

## Column Names

Mengubah nama kolom agar lebih rapi.

```python
new_columns = {
    'Survived': 'SURVIVED',
    'Pclass': 'PCLASS',
    'Name': 'NAME',
    'Sex': 'SEX',
    'Age': 'AGE',
    'Siblings/Spouses Aboard': 'SIBSA',
    'Parents/Children Aboard': 'PARCA',
    'Fare': 'FARE'
}

titanic = titanic.rename(columns=new_columns)
titanic.head()
```

## Replace Values

Mengubah nilai pada kolom `SURVIVED` dan `SEX` agar lebih mudah dibaca.

```python
titanic['SURVIVED'] = titanic['SURVIVED'].replace({0: 'Survived',
                                                   1: 'Not Survived'})
titanic['SEX'] = titanic['SEX'].replace({'male': 'Male',
                                         'female': 'Female'})
```

## Data Types

Menyesuaikan tipe data pada setiap kolomnya.

```python
titanic.dtypes
```

- Kolom `SURVIVED`, `PCLASS`, dan `SEX` diubah menjadi `category` karena merupakan data kategorikal.
- Kolom `AGE` diubah menjadi `int64` karena bilangan bulat.

```python
titanic[['SURVIVED', 'PCLASS', 'SEX']] = titanic[['SURVIVED', 'PCLASS', 'SEX']].astype('category')
titanic['AGE'] = titanic['AGE'].astype('int64')
```

Berikut tipe data setelah diubah:

```python
titanic.dtypes
```

# Data Visualization

Yuk, kita lakukan visualisasi data untuk menggali insight pada `titanic`.

## Single Variable Visualization

### Q: Bagaimana proporsi status keselamatan (`SURVIVED`) penumpang kapal Titanic?

> Lebih banyak penumpang yang selamat (sekitar 550) daripada yang tidak (sekitar 350).

```python
titanic['SURVIVED'].value_counts().plot.bar()
plt.xticks(rotation=0)
plt.xlabel("Survival Status")
plt.ylabel("Number of Passengers")
plt.title("PROPORTION OF SURVIVED VS NOT SURVIVED PASSENGERS", fontweight='bold')
plt.show()
```

### Q: Bagaimana distribusi umur (`AGE`) penumpang yang naik kapal Titanic?

> Cukup banyak penumpang usia 20-30 tahun. Secara keseluruhan bentuk distribusi menyerupai distribusi normal (berbentuk lonceng).

```python
sns.displot(titanic, x='AGE', kind='kde', fill=True)
plt.title("DISTRIBUTION OF PASSENGERS' AGE", fontweight='bold')
plt.show()
```

### Q: Bagaimana distribusi biaya tiket (`FARE`) penumpang?

> Ternyata secara mayoritas (50% data), kisaran biaya tiket hanya berada di range 8-31. Biaya di atas 75 sudah dapat dikategorikan sebagai outlier, dan bahkan ada satu penumpang yang sampai membayar di atas 500.

```python
sns.boxplot(data=titanic, x='FARE')
plt.title("DISTRIBUTION OF PASSENGERS' FARE", fontweight='bold')
plt.show()
```

```python
titanic['FARE'].describe()
```

## Two Variables Visualization

### Q: Bagaimana proporsi status keselamatan (`SURVIVED`) berdasarkan jenis kelamin (`SEX`) penumpang kapal Titanic?

> Ternyata mayoritas penumpang Male selamat, sedangkan untuk Female mayoritas tidak selamat.

```python
titanic.groupby(['SURVIVED', 'SEX'])['NAME'].count().unstack(0).plot.bar(stacked=True)
plt.xticks(rotation=0)
plt.xlabel("SEX")
plt.ylabel("Number of Passengers")
plt.title("PASSENGERS' SURVIVAL STATUS BY SEX", fontweight='bold')
plt.show()
```

### Q: Bagaimana perbedaan distribusi umur (`AGE`) penumpang berdasarkan status keselamatan (`SURVIVED`)?

> Terdapat "punuk" di angka 0-10 pada penumpang yang tidak selamat, mengindikasikan lebih banyak balita dan anak kecil yang tidak selamat dibandingkan yang selamat :(

```python
sns.displot(titanic, x='AGE', hue='SURVIVED', kind='kde', fill=True)
plt.title("DISTRIBUTION OF PASSENGERS' AGE BY SURVIVAL STATUS", fontweight='bold')
plt.show()
```

### Q: Bagaimana perbedaan distribusi biaya tiket (`FARE`) penumpang berdasarkan kelasnya (`PCLASS`)?

> Urutan kelas berdasarkan fare-nya adalah kelas 1, 2, dan 3. Pada kelas 1, biaya tiket cukup bervariasi dibandingkan dua kelas lainnya.

```python
sns.boxplot(data=titanic,
            x='PCLASS',
            y='FARE')
plt.title("DISTRIBUTION OF PASSENGERS' FARE BY PCLASS", fontweight='bold')
plt.show()
```

### Q: Apakah ada korelasi antara umur (`AGE`) dengan biaya tiket (`FARE`)?

> Hampir tidak ada korelasi antara umur dengan biaya tiket.

```python
age_fare_corr = round(titanic.corr().loc['AGE', 'FARE'], 3)

sns.scatterplot(data=titanic,
                x='AGE',
                y='FARE')
plt.title(f"RELATION OF AGE VS FARE (CORRELATION: {age_fare_corr})", fontweight='bold')
plt.show()
```

### Q: Bagaimana proporsi status keselamatan (`SURVIVED`) berdasarkan banyaknya pendamping (`COMPANION`) ?

> Cukup menarik, ternyata apabila naik sendirian atau dengan jumlah pendamping >= 4 lebih besar proporsi yang selamat. Sedangkan dengan jumlah pendamping 1-3 lebih banyak yang tidak selamat.

```python
titanic['COMPANION'] = titanic['PARCA'] + titanic['SIBSA']

companion_survived = titanic.groupby(['COMPANION', 'SURVIVED'])['NAME'].count().unstack(1)
companion_survived.div(companion_survived.sum(axis=1), axis=0).plot.bar(stacked=True)

plt.axhline(0.5, color='r')
plt.legend(bbox_to_anchor=(1, 1))

plt.xlabel("Number of Companions")
plt.ylabel("Percentage")
plt.title("SURVIVAL STATUS PROPORTION BY COMPANION", fontweight='bold')
plt.show()
```

## Multi Variables Visualization

### Q: Bagaimana perbedaan distribusi umur (`AGE`) penumpang berdasarkan jenis kelamin (`SEX`) dan status keselamatan (`SURVIVED`)?

```python
sns.violinplot(data=titanic,
               x='SEX',
               y='AGE',
               hue='SURVIVED')

plt.legend(loc='upper left')
plt.title("DISTRIBUTION OF AGE BY SEX AND SURVIVAL STATUS", fontweight='bold')
plt.show()
```

### Q: Bagaimana nilai korelasi untuk semua variabel numerik pada data `titanic`?

```python
sns.heatmap(titanic.drop('COMPANION', axis=1).corr(), annot=True,
            cmap='RdBu')
plt.show()
```
